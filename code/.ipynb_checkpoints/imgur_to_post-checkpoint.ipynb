```python
from os.path import join
from PIL import Image
import requests
from io import BytesIO

def to_markdown(url, title, center=True):
    
    title_md = '[{:s}]'.format(title)
    url_md = '({:s})'.format(url)
    markdown = '!{:s}{:s}'.format(title_md, url_md)
    
    if center:
        markdown += '{: .center-block :}'

    return markdown + '\n\n'

def load_img(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    return img
    
```

```python
p = 'imgur_urls/nofilter.txt'
```

```python
with open(p, 'r') as file:
    urls = [line.strip() for line in file.readlines()]
    
    def check(url):
        if url[-4:].lower() != '.jpg':
            url += '.jpg'
        return url
    
    urls = [check(url) for url in urls]
```

```python
markdown = ''
for url in urls:
    markdown += to_markdown(url, 'TITLE')
print(markdown)
```
