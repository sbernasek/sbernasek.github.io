```python
def set_location(metadata, gps):
    metadata.exif.gps_latitude = dd_to_dms(gps[0])
    metadata.exif.gps_longitude = dd_to_dms(gps[1])
    del metadata.exif.gps_altitude
    del metadata.exif.gps_altitude_ref
    del metadata.exif.gps_timestamp
    del metadata.exif.gps_datestamp

def save(metadata):
    with open(metadata.path, 'wb') as file:
        file.write(metadata.exif.get_file())  
        
def dd_to_dms(x):
    d = float(int(x))
    m = float(int((x-d)*60))
    s = (x-d-m/60)*3600
    return (d,m,s)    
```
